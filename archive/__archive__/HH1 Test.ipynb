```python
import networkx as nx
import multitool as mt

M = mt.load("HH_61")
c1,c2,c3 = mt.getclust(M,kmeans_clusters = 10, mds_dim = 2, plots = False, layout = nx.circular_layout)
print(c1)
print(c3)
```

```python
M.plot_flat(node_color = c1, layout = nx.kamada_kawai_layout, edge_color = 'gray', node_size = 100, width = 1)
```

<span style='font-size:large'>`Here is the modularity for each layer. Notice most layers have a modularity close to 0.`</span>


```python
print('The modularity of c1 by layer')
print(mt.layer_modularity(M, c1))
print('The modularity of c3 by layer')
print(mt.layer_modularity(M, c3))
```

<span style='font-size:large'>`Here is the modularity for the weighted flattened network. Notice the modularity is still close to 0.`</span>  



```python
print('The modularity of c1 on the flattened network')
print(mt.flatten_modularity(M, c1, weight_count = True))
print('The modularity of c3 on the flattened network')
print(mt.flatten_modularity(M, c3, weight_count = True))
```
